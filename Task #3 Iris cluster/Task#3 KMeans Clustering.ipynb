## PREDICTING OPTIMUM NUMBER OF CLUSTERS IN 'IRIS' DATASET

## Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Reading the data

```python
irs = pd.read_csv("Iris.csv")
irs.head()
```

## Visualizing our data

```python
par=irs.iloc[:,1:]   # Taking every column except ID column
```

```python
from pandas.plotting import parallel_coordinates
plt.figure(figsize=(9,6))
parallel_coordinates(par,'Species', colormap=plt.get_cmap("Set2"))
plt.show()
```

```python
import seaborn as sns
sns.boxplot( x=par["Species"], y=par["SepalLengthCm"], palette="Blues")
```

```python
iris = irs.drop(['Id'],axis=1)  #dropping the Id column as it is not required for classification
iris_df = iris.iloc[:,:-1]
iris_df.head()
```

```python
iris_df.info()
```

## Finding the optimum number of clusters for KMeans 

```python
x = iris_df.values

from sklearn.cluster import KMeans
wcss = []

for i in range(1,11):
    kmeans = KMeans(n_clusters = i, init = "k-means++",max_iter = 300,
                    n_init=10, random_state=42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

## Plotting the results into a line graph and seeing 'the elbow'

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,11),wcss)
plt.title("The elbow method")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()
```

## The elbow point is at 3. So optimum number of clusters is taken as 3

## Applying KMeans to the dataset

```python
kmeans = KMeans(n_clusters = 3, init = "k-means++", max_iter = 300, n_init = 10, random_state = 42)
y_kmeans = kmeans.fit_predict(x)
y_kmeans
```

## Visualising the clusters on first two features(sepal length and sepal width)

```python
plt.figure(figsize=(10,6))
plt.scatter(x[y_kmeans==0,0],x[y_kmeans==0,1], s = 100, c = 'red', label = 'Iris-versicolour')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1],s = 100, c = 'blue', label = 'Iris-setosa')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1],s = 100, c = 'green', label = 'Iris-virginica')

# cluster centers
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 100, c = 'yellow', label = 'Centroids')
plt.legend()
```
